```python
from __future__ import print_function, division

import pandas as pd
import numpy as np

import requests
from lxml import etree
from pattern import web
from urllib import urlretrieve
```

```python
con = requests.get('https://icml.cc/Conferences/2018/Schedule')
```

```python
ele = web.Element(con.content)
```

```python
papers = ele.by_tag('div.maincard narrower Oral')
```

```python
def download(paper):
    try:
        paper_title = str(paper.by_tag('div')[2].content)
        paper_genre = str(paper.by_tag('div')[3].by_tag('a')[0].content.strip())
        paper_page = str(paper.by_tag('div')[5].by_tag('a')[1].attr['href'])
        paper_url = paper_page.replace('.html', '/'+paper_page.split('.html')[0].split('/')[-1]+'.pdf')
    except:
        print('Grabbing paper info Error')
    try:
        download_name = '%s--%s.pdf'%(paper_genre, paper_title)
        done = urlretrieve(paper_url, '/Users/endi/Downloads/icml2018/pdf/%s'%download_name)
    except:
        print('Download Error')
```

```python
for i, paper in enumerate(papers):
    print('================%dth================'%i)
    download(paper)
```
